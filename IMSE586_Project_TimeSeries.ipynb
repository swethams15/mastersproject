# Testing whether the data is stationary

```python
#Method 1
#weather = pd.read_csv('weather_train.csv', header = 0, index_col = 0)

weather['air_temp'] =weather.air_temperature +40

weather['log_air_temperature'] = np.log(weather.air_temp)

weather['log_air_temp_diff'] = weather.log_air_temperature.diff(1)

weather['stationary'] = weather.log_air_temp_diff.diff(periods = 12)



X = weather.groupby('site_id').stationary
```

```python
plt.figure(figsize = (15, 5))

X.hist()## By plotting the histogram we will know that if a bell curve exists then it is a normal distribution. 

plt.xlabel('Air temperature values', fontsize = 20)

plt.ylabel('Site_id in thousands', fontsize = 20)

plt.title('Histogram for air temperature value', fontsize = 20)

plt.show()
```

```python
plt.figure(figsize = (20, 8))

weather.log_air_temperature.plot()

plt.title('Time series plot for stabilizing the variance in the data', fontsize = 20)

plt.xlabel('Time', fontsize = 20)

plt.ylabel('log values of air temperature', fontsize = 20)

plt.show()
```

```python
plt.figure(figsize = (20, 8))

weather.log_air_temp_diff.plot()

plt.title('Time series plot for detrending the data', fontsize = 20)

plt.xlabel('Time', fontsize = 20)

plt.ylabel('Differencing of air temperature', fontsize = 20)

plt.show()
```

```python
## The line plot shows that it is a non stationary data.

plt.figure(figsize = (15, 5))

X.plot()

plt.xlabel('Time', fontsize = 20)

plt.ylabel('Stationary data', fontsize = 20)

plt.title('Line plot for air temperature values.', fontsize = 20)

plt.legend()##-----------Site_id's

plt.show()
```

# Moving Average of air temperature values

```python
r = weather.rolling(window = 12)
r
```

```python
weather['airtemp_ma'] = r.air_temp.mean()

weather[['air_temp', 'airtemp_ma']].head(15)
```

```python
fig = plt.figure(figsize = (20, 8))

plt.style.use('classic')

weather.air_temp.plot(label = 'Positive air temperature values', color = 'red')

weather.airtemp_ma.plot(label = 'Moving Average of air temperature', color = 'grey')

plt.title('Time series plot of air temperature and moving average of air temperature values for 12 months', fontsize = 20)

plt.xlabel('Time', fontsize = 20)

plt.ylabel('Air temperature values', fontsize = 20)

plt.legend()

plt.show()
```

# Seasonal decompose

```python
import statsmodels.api as sm

s = sm.tsa.seasonal_decompose(weather.air_temp, model = 'multiplicative', freq = 12)

fig = plt.figure(figsize = (20, 8))

s.trend.plot(label = 'Trend')

plt.xlabel('Time', fontsize = 20)

plt.ylabel('Air temperature values', fontsize = 20)

plt.title('Line plot for air temperature values showing trends.', fontsize = 20)

fig = plt.figure(figsize = (20, 8))

s.seasonal.plot()

plt.xlabel('Time', fontsize = 20)

plt.ylabel('Air temperature values', fontsize = 20)

plt.title('Line plot for air temperature adjusting the seasonality.', fontsize = 20)

fig = plt.figure(figsize = (20, 8))

s.resid.plot()

plt.xlabel('Time', fontsize = 20)

plt.ylabel('Air temperature values', fontsize = 20)

plt.title('Residual line plot for air temperature.', fontsize = 20)
```

# ACF

```python
import statsmodels.api as sm 

fig = plt.figure(figsize = (20, 8))

plt.style.use('classic')

weather.stationary.dropna(inplace = True)

sm.graphics.tsa.plot_acf(weather.stationary, lags =30, zero = False)

plt.xlabel('Lags', fontsize = 16)

plt.ylabel('Stationary Data', fontsize = 16)

plt.title('Auto Correlation coefficient plot of air temperature', fontsize = 16)

plt.show()
```

```python
#Partial autocorrelation function

fig = plt.figure(figsize = (20, 8))

plt.style.use('classic')

sm.graphics.tsa.plot_pacf(weather.stationary, lags =30, zero = False)

plt.xlabel('Lags', fontsize = 16)

plt.ylabel('Stationary Data', fontsize = 16)

plt.title('Partial Auto Correlation coefficient plot for air temperature', fontsize = 16)

plt.show()
```

# AR Model

```python
from statsmodels.tsa.ar_model import AR

weather_day = weather.air_temp.resample('D').mean()
weather_day.head(60)
```

```python
t= -12
U = weather_day.dropna()
train = U[:t]
test =  U[t:]

fig = plt.figure(figsize = (20, 8))

model = AR(train)
model_fitted = model.fit()
predictions = model_fitted.predict(start = len(U[:t]), end = len((U)-1))
#predictions
predicted = pd.Series(predictions, index = weather_day.fillna(0)[t:].index, name = 'predicted')
predicted.plot()
test.plot()
plt.xlabel('Time', fontsize = 20)
plt.ylabel('Values of the air temperature', fontsize = 20)
plt.title('AR model', fontsize = 20)
print(model_fitted.k_ar)##------p = 16
print(predictions)
plt.legend()
```

```python
from statsmodels.tsa.arima_model import ARMA

weather_day = weather.air_temp.resample('D').mean()
weather_day.head(60)
```

```python
t= -12
U = weather_day.dropna()
train = U[:t]
test =  U[t:]

model = ARMA(train, order = (3,3))
model_fitted = model.fit()
predictions = model_fitted.predict(start = len(U[:t]), end = len((U)-1))
#predictions
fig = plt.figure(figsize = (20, 8))
predicted = pd.Series(predictions, index = weather_day.fillna(0)[t:].index, name = 'predicted')
predicted.plot()
test.plot()
plt.xlabel('Time', fontsize = 20)
plt.ylabel('Values of the air temperature', fontsize = 20)
plt.title('ARMA model', fontsize = 20)
print(model_fitted.k_ar)##------p = 3
print(predictions)
plt.legend()
```
